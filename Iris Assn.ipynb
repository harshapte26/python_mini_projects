```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)

print(df.head())

# Print the information of all attributes of the DS
print(df.info())
```

```python
# Print the statistical information (mean, std, min, etc.) of each feature of the dataset.
print(df.describe())
```

```python
# Print Histograms
plt.hist(df['sepal length (cm)'], bins = 10)
plt.title('Sepal Length')
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count") 
plt.show()

plt.hist(df['sepal width (cm)'], bins = 10)
plt.title('Sepal Width')
plt.xlabel("Sepal Width cm")
plt.ylabel("Count") 
plt.show()

plt.hist(df['petal length (cm)'], bins = 10)
plt.title('Petal Length')
plt.xlabel("Petal Length cm")
plt.ylabel("Count") 
plt.show()

plt.hist(df['petal width (cm)'], bins = 10)
plt.title('Petal Width')
plt.xlabel("Petal Width cm")
plt.ylabel("Count") 
plt.show()
```

```python
# Plotting Boxplot
df.plot.box(title="All Features with outliers")
plt.show()
```
